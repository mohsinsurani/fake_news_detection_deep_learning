```python
import pandas as pd
import re

review_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Reviews.csv')
review_df.columns = review_df.columns.str.strip().str.lower()
review_df['text'] = review_df['text'].map(lambda x: re.sub('[^\w\s]', '', x))
```

```python
review_df.head()
```

```python
from wordcloud import WordCloud
%matplotlib inline
from nltk.corpus import stopwords
from matplotlib import pyplot as plt

stop_words = stopwords.words('english')

most_words = '' 
stopwords = set(stop_words) 
  
for value in review_df['text']: 
    value = str(value) 
    tokens = value.split() 
      
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    most_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 745, height = 745, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 9).generate(most_words)
plt.figure(figsize = (8,8), facecolor = None) 
plt.axis("off") 
plt.imshow(wordcloud) 
plt.show()
```

```python
review_df = review_df[review_df['summary'].notna()]
review_df.dropna(inplace=True)
review_df.drop_duplicates(subset=['summary', 'text'], inplace=True)
```

```python
review_df['len_text'] = review_df['text'].apply(lambda x: len(x))
review_df['outlier'] = review_df['len_text'].apply(lambda x: 1 if (x > 527) else 0)
fin_review_df = review_df[review_df['outlier'] == 0]
fin_review_df['sentiment'] = fin_review_df['score'].apply(lambda x: 1 if (x >= 3) else 0)
df = fin_review_df[['text', 'sentiment']]
```
